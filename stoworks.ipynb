```python
paragraph = """Good morning everyone. Today, I would like to speak about Dr. A.P.J. Abdul Kalam, one of India’s most inspiring leaders. Known as the Missile Man of India, he played a major role in the nation’s space and defence programs. He later became the 11th President of India and was loved as the People’s President. Dr. Kalam strongly believed in the power of dreams, education, and hard work. He inspired millions of young people to think big and serve the nation. His simple life, dedication, humility, and vision continue to guide us. He remains a true symbol of hope and determination."""
```

```python
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
sentences = nltk.sent_tokenize(paragraph)
```

```python
for i  in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words)
```
